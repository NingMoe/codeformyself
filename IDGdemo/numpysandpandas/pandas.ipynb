```python
import pandas as pd

web_stats = {'Day': [1, 2, 3, 4, 5, 6],
             'Visitors': [32, 43, 54, 76, 87, 99],
             'Bounce Rate': [43, 54, 66, 88, 87, 56]
             }
df = pd.DataFrame(web_stats)
# print(df)
# print(df.tail(1))
#当你没有定义索引时，Pandas 会像这样为你生成一个
df.set_index('Day', inplace=True)
# print(df)
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
# print(df['Visitors'])
df['Visitors'].plot()
df['Bounce Rate'].plot()
plt.show()
```
